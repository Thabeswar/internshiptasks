```python
import pandas as pd
```

```python
movies=movies[['id','title','overview','genres',]]
```

```python
movies
```

```python
movies['tags']=movies['overview']+movies['genres']
```

```python
mov=movies.drop(columns=['overview','genres'])
mov=mov.head(10000)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
c=CountVectorizer(max_features=10000,stop_words='english')
```

```python
c
```

```python
vector=c.fit_transform(mov['tags'].values.astype('U')).toarray()
```

```python
vector.reshape(10000,10000)
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
mov[mov['title']=='Toy Story'].index[0]
```

```python
dis=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
```

```python
def recommand(movies):
    ind=mov[mov['title']==movies].index[0]
    dis=sorted(list(enumerate(similarity[ind])),reverse=True,key=lambda vector:vector[1])
    for i in dis[0:10]:
        print(mov.iloc[i[0]].title)
    
```

```python
recommand("Jumanji")
```
